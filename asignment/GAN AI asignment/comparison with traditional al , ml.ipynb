# Traditional AI/ML vs Generative AI Assignment

## Task 1: Comparison Table

| Aspect | Traditional AI / ML | Generative AI (Gen AI) |
|--------|---------------------|------------------------|
| Data Requirements | Requires labeled and structured data for training. | Uses large datasets containing text, images, audio, and other unstructured data. |
| Output Type | Produces predictions, classifications, or recommendations. | Generates new content such as text, images, code, audio, or videos. |
| Human Interaction | Limited interaction; mainly rule-based or predictive. | Supports natural and conversational interactions with users. |
| Adaptability | Optimized for specific tasks and less flexible. | Can adapt to multiple tasks and contexts using prompts. |
| Example Use Cases | Spotify song recommendations, Zomato restaurant recommendations, spam detection. | ChatGPT, Instagram AI-generated filters, DALL-E image generation. |


## Task 2: Zomato Recommendation Scenario

Zomato recommending restaurants based on previous orders mainly uses **Traditional AI/ML**. The system analyzes a user's order history, preferences, ratings, and browsing behavior to predict which restaurants the user is likely to order from next. This is a recommendation problem that relies on predictive analytics and pattern recognition rather than generating completely new content. Therefore, it is an example of traditional machine learning rather than Generative AI.


## Task 3: Improving an App Feature with Gen AI

**Selected Feature: Spotify Song Recommendations**

Traditional AI in Spotify recommends songs by analyzing listening history and user preferences. With Gen AI, Spotify could create personalized playlists along with automatically generated explanations such as *"This playlist is designed for your evening workout mood."* Gen AI could also generate custom music mixes, personalized podcast summaries, or even create AI-generated background music based on the user's mood. This would make the experience more interactive and personalized compared to traditional recommendation systems.


## Task 4: Flipkart Customer Support Chatbot

### Traditional AI/ML Approach
1. Uses predefined rules or intent classification to match user queries with fixed responses.
2. Retrieves answers from an FAQ database and provides standard responses.

### Generative AI Approach
1. Generates natural, context-aware responses instead of selecting only predefined answers.
2. Handles complex or unexpected queries by understanding context and producing flexible, personalized responses.
