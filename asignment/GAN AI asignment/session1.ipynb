## Task 2: Discriminative vs Generative Models

| Aspect        | Discriminative Models                                           | Generative Models                                          |
| ------------- | --------------------------------------------------------------- | ---------------------------------------------------------- |
| Definition    | Learn the boundary between classes and predict labels directly. | Learn the data distribution and generate new data samples. |
| Main Use      | Classification and prediction tasks.                            | Content generation and data creation.                      |
| Example Task  | Spam email detection.                                           | Writing an email or generating images.                     |
| Example Model | Logistic Regression, SVM, Random Forest.                        | GPT, GAN, VAE, Diffusion Models.                           |

## Task 3: Product Description Using ChatGPT

### Prompt Used

"Write a creative product description for a new Flipkart gadget called SmartBand X Pro. The gadget is a lightweight AI-powered smart band with health tracking, sleep monitoring, voice assistant, 10-day battery life, and water resistance. The description should be modern, attractive, and suitable for an e-commerce website."

### Output Received

**SmartBand X Pro – Your Intelligent Fitness Companion**

Take control of your health and productivity with the SmartBand X Pro. Designed for modern lifestyles, this AI-powered smart band tracks your heart rate, steps, sleep quality, and daily activity with exceptional accuracy. Its built-in voice assistant helps you manage reminders, answer questions, and stay organized on the go.

With a sleek lightweight design, 10-day battery life, and water-resistant construction, SmartBand X Pro is built to keep up with every adventure. Whether you're working, exercising, or relaxing, SmartBand X Pro keeps you connected to your wellness goals.

## Task 4: New Generative AI Feature

### Application Chosen: Text Generation

### App: Spotify

### Proposed Feature: AI Music Story Generator

Users can enter a prompt such as "Create a playlist for a road trip with friends." The generative AI system would create a personalized playlist and generate a short story explaining why each song was selected and how it fits the user's mood and activity.

### How Generative AI Powers It

The AI model analyzes user listening history, favorite artists, genres, and the user's prompt. It then generates both a customized playlist and a natural-language explanation, creating a more engaging and personalized music experience.
