
Create a Python class Discriminator with a method discriminate(image) that takes a 28x28 image array and returns a probability score between 0 and 1 indicating if the image is 'real' or 'fake'.<br><br><em><strong>

```python
import numpy as np

class Discriminator:
    def discriminate(self, image):
        """
        Returns a probability score between 0 and 1.
        Uses mean pixel value as a simple rule.
        """
        mean_value = np.mean(image)

        if mean_value > 0.5:
            return 0.8  # likely real
        else:
            return 0.2  # likely fake
```
