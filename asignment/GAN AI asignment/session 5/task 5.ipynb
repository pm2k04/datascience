Task 5: Production Challenges in Generative AI Chatbots


Challenge	:-
Hallucination: Chatbots may provide incorrect or misleading information about bookings or products.	

Solution:-
Use Retrieval-Augmented Generation (RAG) with verified company databases to provide accurate responses.

Challenge	:-
High Cost and Latency: Large AI models can be expensive and slow in production environments.	

Solution :-
Use smaller optimized models, caching, and model quantization to reduce costs and improve response speed.
