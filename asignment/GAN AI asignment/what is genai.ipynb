# Generative AI Assignment

## Task 1: What is Generative AI?

Generative AI (GenAI) is a type of artificial intelligence that can create new content such as text, images, music, videos, and code by learning patterns from existing data. Unlike traditional AI, which mainly analyzes or classifies data, GenAI produces original outputs that resemble human-created content.

**Real-world examples:**
1. **ChatGPT** – Generates human-like responses, answers questions, writes code, and creates content.
2. **DALL-E** – Creates unique images from text prompts provided by users.


## Task 2: Comparison between GenAI and Traditional AI

| Aspect | Generative AI (GenAI) | Traditional AI |
|--------|----------------------|----------------|
| Purpose | Creates new content | Analyzes, predicts, or classifies data |
| Output | Produces original text, images, audio, etc. | Produces fixed outputs such as labels or predictions |
| Creativity | Highly creative and generates new ideas | Limited creativity, follows predefined rules |
| Input Processing | Learns patterns to generate content | Uses patterns mainly for decision-making |
| Examples | ChatGPT, DALL-E, Gemini | Spam filters, recommendation systems, fraud detection |
