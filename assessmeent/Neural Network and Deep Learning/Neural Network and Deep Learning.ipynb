1. What do you understand by Natural Language Processing? 


Natural Language Processing (NLP) is a branch of Artificial Intelligence (AI) that helps computers understand, interpret, process, and generate human language such as English, Hindi, or any other language.

NLP combines:

Computer Science
Artificial Intelligence
Machine Learning
Linguistics

It enables machines to communicate with humans in a natural way.

Applications of NLP
Chatbots
Google Translate
Voice Assistants (Siri, Alexa)
Sentiment Analysis
Spam Detection
Text Summarization
Speech Recognition
Example

When you type:

“I love this product”

An NLP model can identify that the sentence expresses a positive sentiment.



2. What are the steps involved in solving an NLP problem?

The complete step-by-step NLP process is:

Step 1: Data Collection

Collect text data from:

Websites
Social Media
Reviews
Emails
CSV files
Datasets
Example:

Movie reviews dataset from IMDb.

Step 2: Text Cleaning

Remove unwanted data:

Punctuation
Special symbols
HTML tags
Numbers
Extra spaces
Example:

Before:

“Amazing movie!!! ”

After:

“Amazing movie”

Step 3: Tokenization

Split text into smaller parts called tokens (words/sentences).

Example:

Sentence:

“I love NLP”

Tokens:

["I", "love", "NLP"]

Step 4: Stop Word Removal

Remove common words that do not add much meaning.

Stop words:
is
the
and
a
Example:

Before:

“This is a good movie”

After:

“good movie”

Step 5: Stemming / Lemmatization
Stemming

Reduces words to root form.

Example:

playing → play
played → play
Lemmatization

Converts words into meaningful root words.

Example:

better → good

Step 6: Text Vectorization

Convert text into numerical form for machine learning models.

Methods:

Bag of Words (BoW)
TF-IDF
Word2Vec
Embeddings

Step 7: Feature Extraction

Select important text features.

Example:

Frequent words
Word frequency
Sentence length

Step 8: Model Building

Train NLP models using Machine Learning or Deep Learning.

Algorithms:

Naive Bayes
Logistic Regression
SVM
RNN
LSTM
GRU
Step 9: Model Evaluation

Check model performance using:

Accuracy
Precision
Recall
F1-score
Step 10: Deployment

Deploy the model into real-world applications:

Chatbots
Web Apps
Recommendation Systemsb

3. What is an ensemble method in NLP? With Example.

An Ensemble Method is a technique where multiple machine learning models are combined to improve prediction accuracy and performance.

Instead of using one model, ensemble learning uses many models together.

Types of Ensemble Methods

1. Bagging

Combines predictions from multiple models trained independently.

Example:

Random Forest

2. Boosting

Models are trained sequentially to correct previous errors.

Examples:

AdaBoost
XGBoost

3. Voting

Multiple models vote for the final prediction.

Example:

Logistic Regression + SVM + Naive Bayes
Example of Ensemble Method in NLP
Sentiment Analysis Example

Suppose we want to classify movie reviews as:

Positive
Negative

We use:

Naive Bayes
Logistic Regression
Support Vector Machine (SVM)

Each model gives its prediction.

Final prediction is selected using majority voting.

Example:
Model	Prediction
Naive Bayes	Positive
Logistic Regression	Positive
SVM	Negative

Final Output:

Positive

Because 2 out of 3 models predicted Positive.
