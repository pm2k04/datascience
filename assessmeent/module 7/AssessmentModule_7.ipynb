```python

ser = np.random.randint(0, 100, size=20)
minimum = np.min(ser)
q1 = np.percentile(ser, 25)
median = np.median(ser)
q3 = np.percentile(ser, 75)
maximum = np.max(ser)
print("Minimum:", minimum)
print("25th Percentile (Q1):", q1)
print("Median (Q2):", median)
print("75th Percentile (Q3):", q3)
print("Maximum:", maximum)
```

```python
import matplotlib.pyplot as plt
x = [3, 4, 5, 6]
y = [1.5, 2, 2.5, 3]
plt.plot(x, y, marker='o')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Line Plot of Y vs X")
plt.grid(True)
plt.show()
```

```python

x = np.arange(0, 6.1, 0.3)
plt.plot(x, x, "ro", label="x")
plt.plot(x, x**2, "bs", label="x²")
plt.plot(x, x**3, "g", label="x³")
plt.plot(x, x**4, ":", label="x⁴")
plt.xlim(0, 6)
plt.ylim(0, 125)
plt.xlabel("x")
plt.ylabel("y")
plt.title("x, x², x³, x⁴ on the same plot")
plt.legend()
plt.grid(True)
plt.show()
```

```python

height = [179, 155, 191, 152, 188, 177]
names = ['QA', 'WB', 'EC', 'RD', 'TE', 'YF']
plt.bar(names, height, color='skyblue')
plt.title("Height Comparison")
plt.xlabel("Initials")
plt.ylabel("Height (cm)")

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python

x = np.random.randn(100000)
plt.figure(figsize=(15, 4))
plt.subplot(1, 3, 1)
plt.hist(x, bins=10, color='skyblue', edgecolor='black')
plt.title('Histogram with 10 Bins')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.subplot(1, 3, 2)
plt.hist(x, bins=20, color='salmon', edgecolor='black')
plt.title('Histogram with 20 Bins')
plt.xlabel('Value')
plt.subplot(1, 3, 3)
plt.hist(x, bins=50, color='lightgreen', edgecolor='black')
plt.title('Histogram with 50 Bins')
plt.xlabel('Value')
plt.tight_layout()
plt.show()
```
