Question 2: A colleague suggests using a Variational Autoencoder (VAE), while another
recommends a Stable Diffusion-based model. Compare the two approaches for this
image-generation task — address their key architectural differences, output quality, and which is
more appropriate for a production marketing pipeline.


For generating high-quality promotional food images from a restaurant's text description and a reference photo, a Stable Diffusion model is a much better choice than a Variational Autoencoder (VAE).



| Feature                 | Variational Autoencoder (VAE)                                                    | Stable Diffusion                                                                                                          |
| ----------------------- | -------------------------------------------------------------------------------- | ------------------------------------------------------------------------------------------------------------------------- |
| **Architecture**        | Encoder compresses the image into a latent space, and a decoder reconstructs it. | Uses a diffusion process that starts with random noise and gradually generates an image based on text and image guidance. |
| **Input**               | Mainly image data. Limited text conditioning.                                    | Supports both text prompts and reference images (image-to-image generation).                                              |
| **Image Quality**       | Produces smooth but often blurry images.                                         | Produces highly realistic, sharp, and detailed images.                                                                    |
| **Creativity**          | Limited ability to generate new variations.                                      | Can generate many creative and photorealistic variations.                                                                 |
| **Training Complexity** | Easier and faster to train.                                                      | More computationally expensive but delivers much better results.                                                          |
| **Best Use Cases**      | Image compression, anomaly detection, feature learning, simple image generation. | Marketing images, product photography, art generation, advertising, image editing.                                        |


Question 3: GPT-style decoder-only models and BERT-style encoder-only models are both
available to your team. Which architecture is better suited to each job, and what fundamental
property of each architecture makes it the right choice for that specific task?

1. GPT-Style Decoder-Only Model

Best for: Open-ended question answering and text generation.

Fundamental Property:

GPT uses a decoder-only Transformer with causal (masked) self-attention.
It predicts the next word based on the previous words.
This allows it to generate fluent and coherent responses.

Example:
User: "What cuisines are trending near me this weekend?"

GPT Response:

"This weekend, Italian, Chinese, and Mexican cuisines are trending in your area. Many users are also ordering burgers and healthy bowls because of weekend offers."

Why GPT?

Generates complete, human-like answers.
Understands conversational context.
Ideal for chatbots and AI assistants.


2. BERT-Style Encoder-Only Model

Best for: Customer review classification.

Fundamental Property:

BERT uses an encoder-only Transformer with bidirectional attention.
It reads the entire sentence at once (both left and right context).
This helps it understand the full meaning of the text before making a prediction.

Example Review:

"The food arrived late, but the taste was amazing."

BERT Output:

Sentiment: Neutral

Why BERT?

Excellent at understanding sentence meaning.
High accuracy for sentiment analysis and text classification.
Faster and more efficient than generative models for classification tasks.

Difference Between Zero-Shot and Few-Shot Prompting

| Feature               | Zero-Shot Prompting                                       | Few-Shot Prompting                                                                            |
| --------------------- | --------------------------------------------------------- | --------------------------------------------------------------------------------------------- |
| **Definition**        | The model performs a task without any examples.           | The model is given a few examples before solving the task.                                    |
| **Training Examples** | No examples are provided in the prompt.                   | A small number of example input-output pairs are included in the prompt.                      |
| **Performance**       | Good for simple or common tasks but may be less accurate. | Usually more accurate because the model learns the expected response style from the examples. |
| **Best Use Case**     | When no labeled examples are available.                   | When only a few labeled examples (2–5) are available.                                         |



Prompt Structure

A well-structured prompt should include:

1.Role/Instruction
2.A few example complaint–response pairs
3.The new customer complaint
4.The expected output format

Question 5: Describe what distinguishes a LangChain agent from a simple LLM chain. Given that
your food ordering assistant must check live restaurant availability, call an API to calculate
estimated delivery time, and validate discount codes, justify which approach is more
appropriate — and identify two risks the agent approach introduces.


Difference Between a LangChain Agent and a Simple LLM Chain

| Feature             | Simple LLM Chain                                                | LangChain Agent                                                                            |
| ------------------- | --------------------------------------------------------------- | ------------------------------------------------------------------------------------------ |
| **Definition**      | A single LLM call that takes user input and returns a response. | An AI system that can decide which tools or APIs to use before generating a response.      |
| **Decision Making** | No decision-making; follows a fixed workflow.                   | Dynamically decides which action or tool to use.                                           |
| **External Tools**  | Cannot directly use multiple tools or APIs.                     | Can use APIs, databases, search tools, calculators, and other external services.           |
| **Flexibility**     | Best for simple question-answering tasks.                       | Best for complex, multi-step tasks requiring reasoning and external data.                  |
| **Example**         | Answers: "What is pizza?"                                       | Checks restaurant availability, calculates delivery time, validates coupons, then replies. |


Which Approach is More Appropriate?

For the food ordering assistant, a LangChain Agent is the better choice.

Why?

The assistant must perform multiple real-time tasks:

Check live restaurant availability
Calls the restaurant database or API.
Calculate estimated delivery time
Calls a delivery-time API using the customer's location.
Validate discount codes
Calls the coupon validation API.

A simple LLM chain cannot reliably perform these live operations because it only generates text from its training and prompt. A LangChain agent can reason about the user's request, invoke the correct tools or APIs, collect the results, and then generate an accurate response.

Question 6: Compare fine-tuning and RAG for this specific scenario. Address update frequency,
cost, hallucination risk, and data privacy. Which approach is more suitable for a daily-updated,
document-grounded Q&A bot, and what limitations does your recommended approach still
carry?


Comparison of Fine-Tuning and Retrieval-Augmented Generation (RAG)

| Feature                | Fine-Tuning                                                                   | Retrieval-Augmented Generation (RAG)                                                   |
| ---------------------- | ----------------------------------------------------------------------------- | -------------------------------------------------------------------------------------- |
| **Definition**         | Retrains the LLM on domain-specific data to learn new patterns.               | Retrieves relevant documents at query time and uses them as context for the LLM.       |
| **Knowledge Updates**  | Requires retraining whenever data changes.                                    | Simply update the document index; no retraining is needed.                             |
| **Cost**               | High training and maintenance cost.                                           | Lower cost because only document embeddings and indexes need updating.                 |
| **Hallucination Risk** | Higher risk, as the model may generate unsupported information from memory.   | Lower risk because answers are grounded in retrieved documents.                        |
| **Data Privacy**       | Sensitive data becomes part of the trained model, requiring careful handling. | Documents can remain in a secure internal database and are retrieved only when needed. |
| **Best Use Case**      | Stable datasets that rarely change.                                           | Frequently updated knowledge bases and document-based Q&A systems.                     |
